**Matrix Functions**

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

a = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9
b = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9

print("Matrix A:\n", a)
print("Matrix B:\n", b)

# Display matrices
print("The matrix a =\n", a)
print("\nThe matrix b =\n", b)

# Matrix addition
c = a + b
print("\nThe sum of a and b is:\n", c)

# Matrix subtraction
d = a - b
print("\nThe difference of a and b is:\n", d)

# Matrix multiplication
e = np.matmul(a, b)  # or simply 'a @ b'
print("\nThe product of a and b is:\n", e)

# Example matrix A
A = np.array([[1, 2], [3, 4]])

# 1. Size of a matrix
size_A = A.shape  # Returns dimensions (rows, columns)

# 2. Determinant of a matrix
det_A = np.linalg.det(A)  # Returns the determinant of the matrix

# 3. Inverse of a matrix
if np.linalg.det(A) != 0:  # Check if matrix is invertible
    inv_A = np.linalg.inv(A)  # Compute the inverse of A
else:
    inv_A = "Matrix is singular, cannot compute inverse"  # No inverse if determinant is 0

# 4. Rank of a matrix
rank_A = np.linalg.matrix_rank(A)  # Returns the rank of the matrix

# 5. Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)  # Compute eigenvalues and eigenvectors

# 6. Characteristic polynomial
char_poly = np.poly(A)  # Get coefficients of the characteristic polynomial

roots = np.roots(char_poly)



# Display results
print("Size of A:", size_A)
print("Determinant of A:", det_A)
print("Inverse of A:\n", inv_A)
print("Rank of A:", rank_A)

print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
print("Characteristic polynomial coefficients:", char_poly)
print("Roots of the Characteristic Polynomial (Eigenvalues):", roots)

```

```python
A = np.random.randint(1, 15, (4, 4))
B = np.random.randint(1, 15, (4, 4))
print(A)
print(B)

# Flatten the matrices to create sequences for plotting
A_flat = A.flatten()
B_flat = B.flatten()

# Create an index for x-axis
x = np.arange(len(A_flat))
print(x)

# Plot the data
plt.plot(x, A_flat, marker='o', linestyle='-', label="Matrix A")
plt.plot(x, B_flat, marker='s', linestyle='--', label="Matrix B")

# Labels and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Flattened Matrices A and B")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Create two random 4x4 matrices with integer values between 0 and 10
A = np.random.randint(0, 16, size=(4, 4))  # Random integers between 0 and 9
B = np.random.randint(0, 16, size=(4, 4))  # Random integers between 0 and 9

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

# Bar plot to visualize results
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) #figure object, subplot array

# Sum
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout() #plt.tight_layout() is a Matplotlib function that automatically adjusts the layout of subplots and figures to prevent overlapping and ensure a visually appealing presentation. It modifies the spacing between subplots, axes, labels, and titles to improve readability.
plt.show()
```
